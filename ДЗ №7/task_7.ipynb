**Задача №1. [Fibonacci Number](https://leetcode.com/problems/fibonacci-number/description/)**

```python
class Solution:
    def fib(self, n: int) -> int:
        if n < 2:
            return n
        else:
            return self.fib(n-1) + self.fib(n-2)
```

**Задача №2. [Power of Three](https://leetcode.com/problems/power-of-three/description/)**

```python
class Solution:
    def isPowerOfThree(self, n: int) -> bool:
        if n < 1:
            return False
        elif n / 3 == 1 or n == 1:
            return True
        elif n % 3 == 0:
            return self.isPowerOfThree(n // 3)

        return False
```

**Задача №3. [Reverse String](https://leetcode.com/problems/reverse-string/description/)**

```python
class Solution:
    def reverseString(self, s: List[str]) -> None:
        l = 0
        r = len(s) - 1

        s = self.fun(s, l, r)

    def fun(self, string, left, right):
        if left < right:
            string[left], string[right] = string[right], string[left]
            string = self.fun(string, left + 1, right - 1)

        return string
```

**Задача №4. [Binary Search](https://leetcode.com/problems/binary-search/description/)**

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        return self.bin_search(nums, target, 0, len(nums) - 1)


    def bin_search(self, nums, target, left, right):

        mid_index = left + (right - left) // 2

        if nums[mid_index] == target:
            return mid_index
        elif mid_index < left or mid_index > right:
            return -1
            
        if nums[mid_index] > target:
            return self.bin_search(nums, target, left, mid_index - 1)
        else:
            return self.bin_search(nums, target, mid_index + 1, right)
```

**Задача №5. [Search Insert Position](https://leetcode.com/problems/search-insert-position/description/)**

```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        return self.bin_search(nums, target, 0, len(nums))


    def bin_search(self, nums, target, left, right):

        mid_index = left + (right - left) // 2

        if nums[mid_index] == target:
            return mid_index
        elif target > nums[len(nums) - 1]:
            return right
        elif target < nums[0]:
            return left
        elif right - left == 1 and nums[left] < target < nums[right]:
            return right

        if nums[mid_index] > target:
            return self.bin_search(nums, target, left, mid_index)
        else:
            return self.bin_search(nums, target, mid_index, right)
```

**Задача №7. [Peak Index in a Mountain Array](https://leetcode.com/problems/peak-index-in-a-mountain-array/description/)**

```python
class Solution:
    def peakIndexInMountainArray(self, arr: List[int]) -> int:
        return self.peek_index(arr, len(arr) // 2)
    
    def peek_index(self, data, mid):
        if len(data) == 2:
            if data[0] > data[1]:
                return 0
            return 1
        elif len(data) == 1:
            return 0
        elif data[-1] > data[-2]:
            return len(data) - 1
        
        if data[mid] > data[mid - 1] and data[mid] > data[mid + 1]:
            return mid

        if data[mid] > data[mid - 1]:
            return self.peek_index(data, mid + 1)
        else:
            return self.peek_index(data, mid - 1)
```

**Задача №8. [Find Peak Element](https://leetcode.com/problems/find-peak-element/description/)**

```python
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        return self.fpe(nums, len(nums) // 2)
    
    def fpe(self, data, mid):
        if len(data) == 2:
            if data[0] > data[1]:
                return 0
            return 1
        elif len(data) == 1:
            return 0
        elif data[-1] > data[-2]:
            return len(data) - 1
        
        if data[mid] > data[mid - 1] and data[mid] > data[mid + 1]:
            return mid

        if data[mid] > data[mid - 1]:
            return self.fpe(data, mid + 1)
        else:
            return self.fpe(data, mid - 1)
```

**Задача №9. [Search a 2D Matrix](https://leetcode.com/problems/search-a-2d-matrix/description/)**

```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        line = self.search_line(matrix, target, 0, len(matrix))
        
        if line:
            return self.search_column(line, target, 0, len(line))

        return False

    def search_line(self, matrix, target, left, right):
        if target > matrix[-1][-1]:
            return
        elif target < matrix[0][-1]:
            return matrix[0]

        mid = (right + left) // 2

        if matrix[mid][-1] == target:
            return matrix[mid]
        elif left > right:
            return matrix[left]

        if matrix[mid][-1] > target:
            return self.search_line(matrix, target, left, mid - 1)
        else:
            return self.search_line(matrix, target, mid + 1, right)

    def search_column(self, line, target, left, right):
        mid = (right + left) // 2

        if line[mid] == target:
            return True
        elif left > right:
            return False

        if line[mid] > target:
            return self.search_column(line, target, left, mid - 1)
        else:
            return self.search_column(line, target, mid + 1, right)
```
