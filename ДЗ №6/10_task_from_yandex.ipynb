*J. Частотный анализ на минималках:*

```python
unique = set()
letter_count = {}

while True:
    s = input().lower()

    if s != 'финиш':
        for ch in s:
            if ch.isalpha():
                unique.add(ch)

        for ch in unique:
            if ch not in letter_count:
                letter_count[ch] = s.count(ch)
            else:
                letter_count[ch] += s.count(ch)
    else:
        break

max_count = -1
letter = None

for ch in sorted(letter_count):
    if letter_count[ch] > max_count:
        max_count = letter_count[ch]
        letter = ch

print(letter)
```

*O. НОД 3.0:*

```python
numbers = input().split()
nods = []

for i in range(len(numbers) - 1):
    num1 = int(numbers[i])
    num2 = int(numbers[i + 1])
    while num1 != 0 and num2 != 0:
        if num1 >= num2:
            num1 %= num2
        else:
            num2 %= num1
    nods.append(num1 or num2)

print(min(nods))
```

*Q. А роза упала на лапу Азора 5.0:*

```python
s = ''.join(input().split()).lower()
is_palindrome = True

for i in range(len(s) - 1):
    if s[i] != s[len(s) - i - 1]:
        is_palindrome = False

if is_palindrome:
    print('YES')
else:
    print('NO')
```

*R. RLE:*

```python
s = input()
count = 1

for i in range(len(s) - 1):
    if s[i] == s[i + 1]:
        count += 1
    else:
        print(s[i], count)
        count = 1
print(s[-1], count)
```

*F. Кашееды — 3:*

```python
n = int(input())
m = int(input())
first = set()
second = set()

for i in range(n + m):
    inp = input()

    if inp in second:
        first.add(inp)
    else:
        second.add(inp)

sym_diff = first.symmetric_difference(second)

if len(sym_diff) == 0:
    print('Таких нет')
else:
    for surename in sorted(sym_diff):
        print(surename)
```

*H. Кашееды — 4:*

```python
n = int(input())
children = {}

for i in range(n):
    child = input().split()
    
    for el in child[1:]:
        if el in children.keys():
            children[el].append(child[0])
        else:
            children[el] = [child[0]]

need_porridge = input()

if need_porridge not in children.keys():
    print('Таких нет')
else:
    for el in sorted(children[need_porridge]):
        print(el)
```

*J. Транслитерация:*

```python
transliteration = {
    'А': 'A',
    'Б': 'B',
    'В': 'V',
    'Г': 'G',
    'Д': 'D',
    'Е': 'E',
    'Ё': 'E',
    'Ж': 'Zh',
    'З': 'Z',
    'И': 'I',
    'Й': 'I',
    'К': 'K',
    'Л': 'L',
    'М': 'M',
    'Н': 'N',
    'О': 'O',
    'П': 'P',
    'Р': 'R',
    'С': 'S',
    'Т': 'T',
    'У': 'U',
    'Ф': 'F',
    'Х': 'Kh',
    'Ц': 'Tc',
    'Ч': 'Ch',
    'Ш': 'Sh',
    'Щ': 'Shch',
    'Ы': 'Y',
    'Э': 'E',
    'Ю': 'Iu',
    'Я': 'Ia',
    'Ь': '',
    'Ъ': ''
}

text = input()
out = []

for ch in text:
    if ch.upper() not in transliteration:
        out.append(ch)
    else:
        if ch.islower():
            out.append(transliteration[ch.upper()].lower())
        elif ch.isupper():
            out.append(transliteration[ch])

print(''.join(out))
```

*L. Однофамильцы — 2:*

```python
n = int(input())
mans = [input() for i in range(n)]
m_dict = {}

for man in set(mans):
    count = mans.count(man)
    if count > 1:
        m_dict[man] = mans.count(man)

if len(m_dict) == 0:
    print('Однофамильцев нет')
else:
    for man in sorted(m_dict):
        print(f'{man} - {m_dict[man]}')
```

*L. Это будет шедевр!:*

```python
n = int(input())
ingredients = [input() for i in range(n)]
m = int(input())
dishes = {}

for i in range(m):
    dish = input()
    dishes[dish] = []

    for j in range(int(input())):
        dishes[dish].append(input())

count = 0

for dish in sorted(dishes):
    can_cook = True

    for ingredient in dishes[dish]:
        if ingredient not in ingredients:
            can_cook = False

    if can_cook:
        count += 1
        print(dish)

if count == 0:
    print('Готовить нечего')
```
