**Задача A**

```python
if __name__ == '__main__':
    n = int(input())
    words = []

    for i in range(n):
        words.append(input())

    prefix = ('а', 'б', 'в')
    
    count = 0
    
    for i in range(len(words)):
        if words[i].startswith(prefix):
            count += 1
    
    if count == n:
        print('YES')
    else:
        print('NO')
```

**Задача D**

```python
if __name__ == '__main__':
    text = []

    while True:
        inp = input()

        if inp == '':
            break
        elif inp.endswith('@@@'):
            continue
        elif inp.startswith('##'):
            text.append(inp[2:])
        else:
            text.append(inp)

    for el in text:
        print(el)
```

```python
if __name__ == '__main__':
    num = int(input())
    text = []

    for i in range(num):
        text.append(input().split())
    
    counter = 0

    for row in text:
        for col in row:
            if col == 'зайка':
                counter += 1
    
    print(counter)
```

**Задача I**

```python
if __name__ == '__main__':
    text = []

    while True:
        inp = input()
        index = inp.find('#')
        if inp == '':
            break
        elif index == 0:
            continue
        elif index == -1:
            text.append(inp)
        else:
            text.append(inp[:index])

    for el in text:
        print(el)
```

**Задача K**

```python
if __name__ == '__main__':
    n = int(input())
    titles = [input() for i in range(n)]
    search_request = input().lower()

    for title in titles:
        if title.lower().find(search_request) != -1:
            print(title)
```

```python
if __name__ == '__main__':
    kashi = [
        'Манная', 'Гречневая', 'Пшённая', 'Овсяная', 'Рисовая'
    ]

    n = int(input())
    i = 0
    counter = 0
    while counter != n:
        print(kashi[i])

        i += 1
        counter += 1

        if i == len(kashi):
            i = 0
```
