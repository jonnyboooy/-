**Задача №5**

```python
if __name__ == '__main__':
    n = int(input())
    m = int(input())
    first = set()
    second = set()

    for i in range(n + m):
        inp = input()

        if inp in first:
            second.add(inp)
        else:
            first.add(inp)

    sym_diff = first.symmetric_difference(second)
    
    if len(sym_diff) == 0:
        print('Таких нет')
    else:
        print(len(sym_diff))
```

**Задача №8**

```python
if __name__ == '__main__':
    animals = {}

    while inp := input().split():
        for animal in inp:
            if animal not in animals:
                animals[animal] = 1
            else:
                animals[animal] += 1

    for animal in animals:
        print(animal, animals[animal])
```

**Задача №9**

```python
if __name__ == '__main__':
    n = int(input())
    mans_lst = [input() for i in range(n)]
    mans_set = set(mans_lst)
    mans_dict = dict()
    amount = 0

    for el in mans_lst:
        if el not in mans_set:
            mans_set.add(el)

    for el in mans_set:
        buff = mans_lst.count(el)
        if buff != 1:
            mans_dict[el] = buff

    for man in mans_dict:
        amount += mans_dict[man]

    print(amount)
```

**Задача №10**

```python
if __name__ == '__main__':
    inp = input().split()
    nums = []

    for i in range(len(inp)):
        inp[i] = bin(int(inp[i]))[2:]

    for num in inp:
        nums.append({
            "digits": len(num),
            "units": num.count('1'),
            "zeros": num.count('0')
        })
    print(nums)
```
